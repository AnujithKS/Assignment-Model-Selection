```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv('/content/titanic_dataset .csv')
```

```python
dt
```

```python
dt.isna().sum()
```

```python
dt= dt.drop('Cabin',axis=1)
```

```python
dt['Age'].fillna(dt['Age'].mean(),inplace=True)
```

```python
dt['Embarked'].fillna(dt['Embarked'].mode()[0],inplace=True)
```

```python
dt.isnull().sum()
```

```python
y=dt['Survived']
```

```python
x=dt.drop('Survived',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dt['Embarked'] = label_encoder.fit_transform(dt['Embarked'])
dt['Name'] = label_encoder.fit_transform(dt['Name'])
dt['Sex'] = label_encoder.fit_transform(dt['Sex'])
dt['Ticket'] = label_encoder.fit_transform(dt['Ticket'])
```

```python
dt.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state = 42, test_size = 0.2)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
score_knn=knn.score(x_test,y_test)
```

```python
score_knn
```

SVM

```python
from sklearn.svm import SVC
svc= SVC()
svc.fit(x_train, y_train)
```

```python
score_svc=svc.score(x_test,y_test)
```

```python
score_svc
```

Kfold

```python
from sklearn.model_selection import KFold
kfold_validator  = KFold(10)
for train_index,test_index in kfold_validator.split(x,y):
    print('Training Index: ', train_index)
    print('Testing Index: ',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_result = cross_val_score(knn,x,y,cv = kfold_validator)
```

```python
cross_val_result
```

```python
np.mean(cross_val_result)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_result2 = cross_val_score(svc,x,y,cv = kfold_validator)
```

```python
cross_val_result2
```

```python
np.mean(cross_val_result2)
```

**Stratified cross validation**

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator = StratifiedKFold(n_splits = 10)
```

```python
for train_index,test_index in skfold_validator.split(x,y):
    print('Training Index: ', train_index)
    print('Testing Index: ',test_index)
```

```python
scross_val_result1 = cross_val_score(knn,x,y,cv = skfold_validator)
```

```python
scross_val_result1
```
